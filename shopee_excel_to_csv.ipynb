```python
import pandas as pd
import os
from time import time
import warnings

warnings.filterwarnings("ignore", message="Workbook contains no default style, apply openpyxl's default")

folder = 'Tiny Buds 2'
shop = 'Tiny Buds'
sc_id = '18520495'
extract_date = '2024-04-19T16:44:00'

directory = f'Shopee Seller Center/{folder}'
excel_files = os.listdir(directory)
i = 1

for file in excel_files:
    t_start = time()
    print(f"Converting to csv: {file}")
    file_path = os.path.join(directory, file)
    df = pd.read_excel(file_path)
    df.insert(0, 'EXTRACT_DATE', extract_date)
    df.insert(0, 'SC_ID', sc_id)
    csv_file = f"Shopee {shop} (March 24 - April 13) part {i} of {len(excel_files)}.csv"
    df.to_csv(f'S3 Files/end_status_report/spe/{csv_file}', index=False)
    t_end = time()
    print(f'Done converting: {csv_file}')
    print('Took %.2f seconds' % (t_end - t_start))
    print()
    i += 1
```
